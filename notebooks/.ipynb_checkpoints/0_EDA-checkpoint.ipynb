## Bibliotheken

```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
import missingno as msno
```

## Daten laden
daten laden / erstbehnadlung missing values || im orginal ds als "?", ich konvertiere zu np.nan / Ändern datatype alles auf float bis auf normalized loss 

```python
labelList=['symboling','normalizedLosses','make','fuelType','aspiration','numOfDoors','bodyStyle','driveWheels','engineLocation',
           'wheelBase','length','width','height','curbWeight','engineType','numOfCylinders','engineSize','fuelSystem','bore',
           'stroke','compressionRatio','horsepower','peakRpm','cityMpg','highwayMpg','price']
df = pd.read_csv('../data/data_car-CopyForEDA.csv',delimiter=',',encoding='utf-8', names=labelList)
# Datentypen umwandeln & ? durch np.NAN ersetzen
df = df.replace("?", np.nan)
df[['normalizedLosses']]=df[['normalizedLosses']].astype('category') #
df[['price']]=df[['price']].astype('float64')
df[['horsepower']]=df[['horsepower']].astype('float64')
df[['peakRpm']]=df[['peakRpm']].astype('float64')
df[['bore']]=df[['bore']].astype('float64')
df[['stroke']]=df[['stroke']].astype('float64')
df[['engineSize']]=df[['engineSize']].astype('float64')
df[['cityMpg']]=df[['cityMpg']].astype('float64')
df[['highwayMpg']]=df[['highwayMpg']].astype('float64')
df[['curbWeight']]=df[['curbWeight']].astype('float64')
```

## Informationen aus dem DF
es gibt die Dtypen int64, object, float64 und category<br>
Die featurenormalizedLosses, numOfDoors,bore, stroke horsepower,peakRPM haben missings(np.nans) => hier wird auf irgendeine weise imputed werden<br>
die Information aus der Funktion describe ist für mich , dass es auf den ersten Blick (vgl mean, max, min) die Daten für mich nicht großartig mit ausreißern "gespickt sind" => Hilfestellung für die Skalierung

Eine wie ich finde gute Möglcihkeit die NAs eines Datensatzes darzustellen ist via der "missing-nos" Matrix. Diese habe cih im folgenden ausgeführt.

```python
# show all nan-Values
msno.matrix(df)
```

# Korrelationen
Wenden wir auf unser DF nun Funktionnen der Korrelation an, so werden automatisch lediglich die numerischen Feature einbezogen.

```python
df.corr()[["price"]].sort_values(by="price")
```

Uns interessiert wie die Variablen zusammenhängen, dies wird u.a. in der Korrelation ausgedrückt. Die Variable Preis eines Fahrzeugs scheint somit in erster Linie von der engineSize (Hubraum), dem CrubWeight Leergewicht und der Horsepower(PS) abhängig zu sein. Das würde nun bedeuten, dass schwere Fahrzeuge (Bspw SUV) und höher motorisierte Fahzeuge(engine size & HP) grundsäzlich teuere sind. So weit scheint das für mich nachvollziehbar.

```python
df.corr()[["symboling"]].sort_values(by="symboling")
```

betrachten wir nun die Zusammenhänge für die Versicherungseinstufung(symboling) so scheinen die Parameter HP&Peakrpm ("ganz" indirekt somit auch der Verbauch highwaympg) ausschlaggebend zu sein. je gößer der Wert in symboling ist, desto risikoreicher ist die Einstufung, die durch dieses Skala ausgedrückt werden soll. <br>

Nachfolgend habe ich nochmals die nicht gefilterte Korrelation zunächst in Tabellenform und anschließend als clustermap & heatmap dargestellt.


```python
df.corr()
```

```python
sns.clustermap(df.corr())
```

```python
sns.heatmap(df.corr()) 
```

## Darstellung der Fahrzeugpreise nach Bauart(Body-Styyle)


```python
plt.figure(figsize=(20,10))
sns.stripplot(data=df, x='bodyStyle', y='price', jitter=0.3)
```

Hier kann man nun visuell erkennen(ohne entsprechende Python abfragen wie bswp "car_df.BodyStyle.value_counts()"), dass wohl in dem gegeben Datensatz die Fahrzeugtypen "Sedan" & hatchback am häfigsten auftretten. Das teruerst Fahrzeug scheint ein Harttop zu sein un der Fahrzeugtyp convertible steuert die wengisten Instanzen dem Dataframe bei. 

```python
plt.figure(figsize=(20,10))
sns.stripplot(data=df, x='make', y='price', jitter=0.3)
```
